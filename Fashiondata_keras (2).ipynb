```python
from transformers import pipeline
from PIL import Image
import requests
import time

# Image URL or local file
img_url = "https://img.freepik.com/premium-photo/group-diverse-business-people-working-together-having-meeting_695793-1294.jpg?w=2000"
image = Image.open(requests.get(img_url, stream=True).raw)

# List of model IDs to test
model_ids = [
    "google/vit-base-patch16-224",
    "facebook/deit-base-distilled-patch16-224",

    "microsoft/swin-base-patch4-window7-224",

]

# Test each model
results = []

for model_id in model_ids:
    print(f"\n🔍 Testing model: {model_id}")

    try:
        # Create pipeline for each model
        classifier = pipeline("image-classification", model=model_id, device=-1)

        # Time inference
        start = time.time()
        preds = classifier(image, top_k=1)
        end = time.time()

        label = preds[0]['label']
        score = preds[0]['score']
        duration = end - start

        results.append({
            "model": model_id,
            "label": label,
            "score": round(score, 4),
            "time_sec": round(duration, 3)
        })
    except KeyError as e:
        print(f"Skipping model {model_id} due to KeyError: {e}")
        results.append({
            "model": model_id,
            "label": "Error",
            "score": 0.0,
            "time_sec": 0.0
        })
    except Exception as e:
        print(f"Skipping model {model_id} due to unexpected error: {e}")
        results.append({
            "model": model_id,
            "label": "Error",
            "score": 0.0,
            "time_sec": 0.0
        })


# Print summary
print("\n📊 Summary:")
for res in results:
    print(f"{res['model']:<50} | {res['label']:<20} | Score: {res['score']} | Time: {res['time_sec']}s")
```

```python
from transformers import pipeline
from PIL import Image, ImageDraw, ImageFont
import requests
import matplotlib.pyplot as plt

# Step 1: Load image
image_url = "https://img.freepik.com/premium-photo/group-diverse-business-people-working-together-having-meeting_695793-1294.jpg?w=2000"
image = Image.open(requests.get(image_url, stream=True).raw).convert("RGB")

# Step 2: Load object detection model
detector = pipeline("object-detection", model="facebook/detr-resnet-50", device=-1)

# Step 3: Run detection
results = detector(image)

# Step 4: Draw boxes
draw = ImageDraw.Draw(image)
font = ImageFont.load_default()

for obj in results:
    box = obj["box"]
    label = obj["label"]
    score = obj["score"]
    draw.rectangle(
        [(box["xmin"], box["ymin"]), (box["xmax"], box["ymax"])],
        outline="red", width=2
    )
    draw.text((box["xmin"], box["ymin"] - 10), f"{label} ({score:.2f})", fill="red", font=font)

# Step 5: Show image with boxes
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis("off")
plt.show()
```
